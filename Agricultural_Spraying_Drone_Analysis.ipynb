```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

This generates synthetic farm dataset.

```python
# Generate synthetic farm data

n = 200

data = pd.DataFrame({
    "Crop_Type": np.random.choice(["Wheat","Rice","Cotton","Soybean"], n),
    "Field_Size_Acres": np.random.randint(1,10,n),
    "Crop_Health_Index": np.random.uniform(0.5,1.0,n),
    "Pesticide_Usage_Liters": np.random.uniform(2,8,n),
    "Drone_Flight_Time_Minutes": np.random.randint(5,20,n),
    "Spraying_Efficiency": np.random.uniform(80,98,n)
})

data.head()
```

This shows:

average pesticide usage

average spraying efficiency

field size statistics

**What agricultural and drone variables should be collected? Explain their importance.**

Agricultural and Drone Variables

To implement an AI-enabled agricultural spraying drone system, several important variables must be collected.

**Agricultural Variables**

Crop Type – Different crops require different pesticides and spraying techniques.

Field Size – Helps determine the drone flight path and the amount of pesticide required.

Crop Health Index – Indicates the condition of crops and helps identify areas needing pesticide treatment.

Pesticide Usage – Helps track how much pesticide is applied and ensures efficient usage.

**Drone Variables**

Drone Flight Time – Determines how long the drone operates for spraying tasks.

Spraying Efficiency – Measures how effectively pesticides are applied across the field.

GPS Navigation – Helps the drone move accurately across the farm and follow optimized flight paths.

These variables allow the system to optimize pesticide spraying, reduce chemical waste, and improve overall farming efficiency.

```python
data.describe()
```

This shows which crops are most common

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Crop_Type", data=data)

plt.title("Crop Type Distribution")
plt.xlabel("Crop Type")
plt.ylabel("Number of Fields")

plt.show()
```

Shows how efficient drone spraying is.

```python
plt.figure(figsize=(6,4))

plt.hist(data["Spraying_Efficiency"], bins=10)

plt.title("Spraying Efficiency Distribution")
plt.xlabel("Efficiency (%)")
plt.ylabel("Frequency")

plt.show()
```

Shows that larger fields require more pesticide.

```python
plt.figure(figsize=(6,4))

plt.scatter(data["Field_Size_Acres"], data["Pesticide_Usage_Liters"])

plt.title("Field Size vs Pesticide Usage")
plt.xlabel("Field Size (Acres)")
plt.ylabel("Pesticide Usage (Liters)")

plt.show()
```

**Average Spraying Efficiency**

Meaning drones spray very efficiently.

```python
avg_efficiency = data["Spraying_Efficiency"].mean()

print("Average Spraying Efficiency:", avg_efficiency)
```
